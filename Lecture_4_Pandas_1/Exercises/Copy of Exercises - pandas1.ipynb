## Exercise 3. DataFrame

Create a numpy array of random integers between 200 and 300, with size (20, 3)

```python
arr = np.random.randint(200, 301, size=(20,3))
arr
```

Create a DataFrame from the previous array, and name the columns A, B and C.
Then print its shape, and show its head

```python
df = pd.DataFrame(arr, columns=['A','B','C'])
print(df.shape)
df.head()
```

Repeat the previous question (with the same array), but this time use as index the numbers from 25 to 45

```python
df = pd.DataFrame(arr, columns=['A','B','C'], index=np.arange(25,45))
print(df.shape)
df.head()
```

Rename the column A to Q, and column C to Z. (show the head to see the changes)

```python
df.columns = ['Q','B','Z']
```

```python
df.rename(columns={'A':'Q', 'C':'Z'}, inplace=True)
df
```

Make a numpy array out of the DataFrame. 
Then, reshape that array into (12,5), and create a new DataFrame out of it with column names U,V,W,X,Y

```python
a = df.to_numpy()
print(a.shape)
a = a.reshape(12, 5)
df2 = pd.DataFrame(a, columns=['U','V','W','X','Y'])
print(df2.shape)
df2.head()
```

Save the DataFrame into a csv file (named `my12x5df.csv`) (excluding the index)

```python
df2.to_csv('my12x5df.csv', index=False)
```
