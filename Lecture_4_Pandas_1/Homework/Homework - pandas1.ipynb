## Question 2.

We have three lists:
* the 1st contains the names of the largest cities in Germany, 
* the 2nd contains the states to which those cities belong (in order),
* the 3rd contains the population of those cities (in order).

```python
cities = ['Berlin','Hamburg','Munich','Cologne','Frankfurt',
          'Stuttgart','Dusseldorf','Dortmund','Essen','Leipzig']
states = ['Berlin','Hamburg','Bavaria','NRW','Hesse',
          'Baden-Wurttemberg','NRW','NRW','NRW','Saxony']
pops = [3520031, 1787408, 1450381, 1060582, 732688, 
        623738, 612178, 586181, 582624, 560472]
```

**Q2.1.** Create a dictionary `dic` whose keys are `'City'`, `'State'`, `'Population'` and the associated values are the three lists, respectively.

```python
# write your code here
dic = {'City':cities,'State': states,'Population': pops}
print(dic)
```

**Q2.2.** Create a DataFrame from the dictionary. Print its shape and show it (the entire dataframe)

```python
# write your code here
df = pd.DataFrame(dic)
print(df.shape)
df
```

**Q2.3.** 
* Instead of creating a dictionary, use `zip()` to combine the three lists into a list of tuples.
* Then, create a DataFrame from it (don't forget column names).
* You should get the same dataframe

```python
# write your code here
data = zip(cities,states,pops)
df2 = pd.DataFrame(data, columns = ['City','State','Population'])
print(df2.shape)
df2
```

**Q2.4.** Change the column names into `Stadt`, `Staat`, `Einwohnerzahl`

```python
# write your code here
df2.columns=['Stadt','Staat','Einwohnerzahl']
df2
```

**Note**

This dataset contains information about the passengers of the Titanic, the famous ship that sank in 1912.

The data consists of the following columns:

* `PassengerId`: Id of every passenger.
* `Survived`: This feature have value 0 and 1. 0 for not survived and 1 for survived.
* `Pclass`: There are 3 classes: Class 1, Class 2 and Class 3.
* `Name`: Name of passenger.
* `Sex`: Gender of passenger.
* `Age`: Age of passenger.
* `SibSp`: Indication that passenger have siblings and spouse.
* `Parch`: Whether a passenger is alone or have family.
* `Ticket`: Ticket number of passenger.
* `Fare`: Indicating the fare.
* `Cabin`: The cabin of passenger.
* `Embarked`: The embarked category.
