## Exercise 5. Data Merging

A company is selling some tech products. 

They hire **you** as a **Data Scientist**.

In your first work day, they told you that their data is stored in several csv files, and they assigned to you several tasks.

**Data Files**
1. `items.csv` describes items, each item has: `item_id`, `item_name`, `brand`, `unit_price`
2. `transactions.csv` describes the selling transaction, each transaction has: `trans_id`, `customer_id`, `month`, `year`
3. `transaction_details.csv` provides details of transactions, i.e., what are the items of that transaction and the quantity of each: `transaction_id`, `item_id`, and `quantity`.

**Tasks** 
1. Find the total price of each transaction
2. Find the total sales (sum of transaction prices) per month
3. Find the total sales per item
4. Find the total sales per customer

So, your first step is to load the data and explore it.

Read the three csv files into DataFrames (print the shape and show the head):
* `items.csv` into `df_items`
* `transactions.csv` into `df_trans`
* `transaction_details.csv` into `df_details`

```python
# read `items.csv`
df_items = pd.read_csv('items.csv')
print(df_items.shape)
df_items.head()
```

```python
# read `transactions.csv`
df_trans = pd.read_csv('transactions.csv')
print(df_trans.shape)
df_trans.head()
```

```python
# read `transaction_details.csv`
df_details = pd.read_csv('transaction_details.csv')
print(df_details.shape)
df_details.head()
```

**Task 1.** Find the total price of each transaction
* transaction total price = sum of prices of each item of the transaction
* item price = quantity * unit price
    
Notice that we need to take the `quantity` from `transactions_details`, and take the `unit_price` from `items`.

```python
# merge `df_details` with `df_items` (call the result `df`)
df = df_details.merge(df_items, on='item_id')
print(df.shape)
df.head()
```

```python
# then, merge `df` with `df_trans`
df = df.merge(df_trans, left_on='transaction_id', right_on = 'trans_id')
print(df.shape)
df.head()
```

```python
# drop column `trans_id` since it is duplicated
df.drop(columns='trans_id', inplace=True)
```

```python
# calculate the price of each item
df['price'] = df['quantity']*df['unit_price']
df.head()
```

```python
# group by `transaction_id` and sum the prices (call the result `df2`)
df2 = df.groupby('transaction_id').sum()['price'].reset_index()
df2
```

**Task 2.** Find the total sales (sum of transaction prices) per month

Notice that we take the month,year from `transactions`, and the transaction prices from `df2`


```python
# merge `df2` with `df_trans` (call the result `df3`)
df3 = df2.merge(df_trans, left_on='transaction_id', right_on='trans_id')
df3.head()
```

```python
# group by month and year, and sum the price. (then sort by month,year)
df3.groupby(['year','month']).sum(['price']).reset_index().sort_values(by=['year','month'])
```

**Task 3.** Find the total sales per item

```python
df.groupby(['item_name']).sum()['price']
```
