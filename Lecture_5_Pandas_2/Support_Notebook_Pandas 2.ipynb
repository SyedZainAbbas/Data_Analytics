### 1.2. Filter specific rows from a DataFrame

To select rows based on a conditional expression, use a condition inside the selection brackets `[ ]`.

```python
df[df['Age']>30]
```

* The output of the conditional expression is actually a pandas Series of boolean values (either True or False) 
    * with the same number of rows as the original DataFrame.
* This Series of boolean values can be used to filter the DataFrame by putting it in between the selection brackets `[ ]`. 
* Only rows for which the value is True will be selected

```python
df['Age']>30
```

What can we use with conditions:
* Comparison operators: `==`, `!=`, `<`, `<=`, `>`, `>=`
* Logical connectors: `&` , `|` , `~`
* Functions: `.isin()`, `.isna()`, `.isnull()`, `.notna()`

```python
# return people in their 30s
df[(df['Age']>=30) & (df['Age']<40)]
```

```python
# return Male people from Munich
df[(df['Gender']=='Male') & (df['Address']=='Munich')]
```

```python
# return the rows where the address is Hamburg or Berlin
df[(df['Address']=='Hamburg') | (df['Address']=='Berlin')]
```

```python
# return the rows where the address belong to the list [Hamburg, Berlin]
df[df['Address'].isin(['Hamburg', 'Berlin'])]
```

```python
# return the rows where the Age is unknown (missing)
df[df['Age'].isnull()]
```

```python
# return the rows where the Age is known (not missing)
df[~df['Age'].isna()]
```

```python
# or:
df[df['Age'].notna()]
```

#### `loc` and `iloc` operators
* `loc` is label-based selection operator
* we specify rows and columns based on their row and column labels 
    * `df.loc[row_labels, column_labels]`
* `iloc` is position-based selection operator
* we specify rows and columns by their integer position values (0-based)
    * `df.iloc[row_positions, column_positions]`

```python
# give me rows 2 till 4 and columns 3 to 4
# df.iloc[df["Age"] > 30, 3:5] does not work
# df.iloc[2:5, ['FirstName', 'Address']] only takes integers
# df.iloc[list(df["Age"] > 30), 3:5] this works. It does not take the series, we have to first convert the bool series into bool list
df.iloc[2:5, 3:5]
```

* `.map()` is used to apply a mapping (function or dict) element-wise for a single series

* `.apply()` is used to apply a mapping (function) using multiple dataframe columns/rows

```python
corrections = {'maxda': 'mazda', 'vokswagen': 'volkswagen'}
df['car-name'] = df['car-name'].map(corrections)
df
```

This is bad !

Values that are not found in the dict are converted to NaN

So, how can we solve it?!

```python
df = pd.DataFrame(carprices, columns=['car-id','car-name','price'])
```

```python
corrections = {'maxda': 'mazda', 'vokswagen': 'volkswagen'}
df['car-name'] = df['car-name'].map(lambda x: corrections.get(x,x) )
df
```

#### Note about `inplace` argument 
 
Many pandas methods accepts the argument inplace.
* If `inplace` is set to `True`, then the changes that the method causes are done in place, i.e., this will modify the original dataframe.
* If `inplace` is set to `False`, the method returns a new dataframe with updated contents, and the original dataframe is not changed.

By default, `inplace` is `False`. This means:
* by default your changes don’t affect the original dataframe, but creates a new one! 
* If you want to apply your changes permanently to the original dataframe, you can either:
    * use the option: `inplace=True`, i.e., `df.some_function (..., inplace=True)`
    * or assign the result to the same dataframe, i.e., `df = df.some_function(...)`
 

```python
df = pd.DataFrame(np.arange(16).reshape(4,4), 
                  columns=['A','B','C','D'], 
                  index=['e','f','g','h'])
df
```

```python
# This will not change the original df
df.drop(index=['f','g'])
```

```python
df
```

```python
# This will change the original df
df.drop(index=['f','g'], inplace=True)
df
```

### 2.5 Iterate over DataFrames

Iterate over rows
* `iterrows()`


```python
d= [('West Ham', 'Wolves', 1, 0), ('Everton','Man City', 0, 1),
    ('Man United','Watford', 0, 0),  ('Liverpool','Leeds United', 6, 0),
    ('Leeds United','Man United', 2, 4), ('Arsenal', 'Wolves', 2, 1)]
df = pd.DataFrame(d, columns=['Team 1', 'Team 2', 'G1', 'G2'])
df
```

```python
for i,row in df.iterrows():
    print(i,row)
    # print(i, row['Team 1'],  row['Team 2'], row['G1'], row['G2'])
```

```python
# Count the goals of each team
goals_dic = {}
for i, row in df.iterrows():
    team1, team2 = row['Team 1'], row['Team 2']
    goals1, goals2 = row['G1'], row['G2']
    goals_dic[team1] = goals_dic.get(team1,0) + goals1
    goals_dic[team2] = goals_dic.get(team2,0) + goals2
goals_dic  
```

```python
# Make a list of goals of each team
goals_dic = {}
for i, row in df.iterrows():
    team1, team2 = row['Team 1'], row['Team 2']
    goals1, goals2 = row['G1'], row['G2']
    goals_dic[team1] = goals_dic.get(team1,[]) + [goals1]
    goals_dic[team2] = goals_dic.get(team2,[]) + [goals2]
goals_dic  
```

```python
# count the goals of each team
goals_dic = {}
for team1, team2, g1, g2 in df.itertuples(index=None, name=None):
    goals_dic[team1] = goals_dic.get(team1,0) + g1
    goals_dic[team2] = goals_dic.get(team2,0) + g2
goals_dic 
```

```python
from collections import defaultdict
```

```python
goals_dic = defaultdict(lambda : 0)
for team1, team2, g1, g2 in df.itertuples(index=None, name=None):
    goals_dic[team1] += g1
    goals_dic[team2] += g2
goals_dic = dict(goals_dic)
goals_dic
```
