## Question 2.

Read the file `golf.csv` into a dataframe, print the shape and show the head

```python
# write your code here
df = pd.read_csv('golf.csv')
print(df.shape)
df.head()
```

**Q2.1** Convert the column `Play` replacing `yes` by 1, and `no` by 0

```python
# write your code here
# df['Play'].replace({'yes': 1, 'no': 0}, inplace = True)
df['Play'] = df['Play'].map({'yes': 1, 'no': 0})
df
```

**Q2.2.** Convert the column `Windy` replacing True by 1, and False by 0

```python
# write your code here
df['Windy'].replace({True: 1, False: 0}, inplace = True)
df
```

**Q2.3.** Create a new column, call it `x`, whose values are:
* 1: if `Outlook` is `overcast`
* 0: otherwise

```python
# write your code here
df['x'] = df.apply(lambda row: 1 if row['Outlook'] == 'overcast' else 0, axis = 1)
df
```

**Q2.4.** Create a new column, call it `y`, whose values are:
* 1: if `Outlook` is `rainy` and `Windy` is 0
* 0: otherwise

```python
# write your code here
df['y'] = df.apply(lambda row: 1 if (row['Outlook'] == 'rainy') & (row['Windy'] ==0) else 0, axis = 1)
df
```

**Q2.5.** Create a new column, call it `z`, whose values are:
* 1: if `Outlook` is `sunny` and `Humidity` is `normal`
* 0: otherwise

```python
# write your code here
df['z'] = df.apply(lambda row: 1 if (row['Outlook']=='sunny') & (row['Humidity']=='normal') else 0, axis = 1)
df
```

**Q2.6.** Create a new column, call it `a`, whose values are:
* 1: when any of the three new columns `x`, `y`, or `z` is 1 
* 0: otherwise (i.e., all the three columns are 0)

```python
# write your code here
df['a'] = df.apply(lambda row: 1 if (row['x'] == 1) | (row ['y'] == 1) | (row ['z'] == 1) else 0, axis = 1)
df
```

**Q2.7.** Select the rows where the columns `Play` and `a` are not equal;

## Question 3.

Read the `sales.csv` file into a DataFrame, print the shape and show the head

```python
# write your code here
df = pd.read_csv('sales.csv')
print(df.shape)
df.head()
```

In addition, we have the following dictionary which associates each product with a tuple comprising the buy unit price and the sell unit price of that product.

```python
prices = {'Tablet':(420, 450), 
          'Headset':(80, 90), 
          'Printer':(160, 180), 
          'Monitor':(200, 230), 
          'Laptop':(1000, 1100)}
```

**Q3.1.** Create two columns in the DataFrame:
1. `buy_unit_price` which comprises the buy unit price of the product, and
2. `sell_unit_price` which comprises the sell unit price of the product

(Hint: you need to use lambda functions)

```python
# write your code here
df['buy_unit_price'] = df.apply(lambda row: prices[row['product']][0], axis=1)
df['sell_unit_price'] = df.apply(lambda row: prices[row['product']][1], axis=1)
df.head()
```

**Q3.2.** Create other two columns in the DataFrame:

3. `buy_total` which comprises the quantity multiplied by the buy unit price of the product, and
4. `sell_total` which comprises the quantity multiplied by the sell unit price of the product.

```python
# write your code here
df['buy_total'] = df['quantity']*df['buy_unit_price']
df['sell_total'] = df['quantity']*df['sell_unit_price']
df.head()
```

**Q3.3.** Now, create a column named `profit` which is the difference between the sell total and the buy total.

```python
# write your code here
df['profit'] = df['sell_total'] - df['buy_total']
df.head()
```

**Q3.4.** Drop the columns `buy_unit_price` and `sell_unit_price`

```python
# write your code here
df.drop(columns = ['buy_unit_price', 'sell_unit_price'], inplace=True)
df
```

**Q3.5.** Find the sum of profit in year 2020

```python
# write your code here
df['profit'].sum()
```

**Q3.6.** Find the average profit of Tablet products

```python
# write your code here
df.loc[df['product'] == 'Tablet','profit'].mean()
```
