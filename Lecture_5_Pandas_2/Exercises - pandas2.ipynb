## Exercise 4. Data Manipulation

Creat a numpy array of random integers, between 50 and 100, of size (10,5).
Then make a DataFrame out of it, call the columns: `Math`, `Physics`, `English`, `History`, `Geography`.

```python
arr = np.random.randint(50, 101, size=(10,5))
df = pd.DataFrame(arr, columns=['Math', 'Physics', 'English', 'History', 'Geography'])
df
```

Create a new column, called `avgSc`, whose values are a weighted average of Math (60%) and Physics (40%) (round it to integer)

```python
df['avgSc'] = round(df['Math']*.6 + df['Physics']*.4).astype(int)
df
```

Create a new column, called `avgA`, 
whose values are a weighted average of English (40%), History (30%), and Geography (30%) (round it to integer)

```python
df['avgA'] = df.apply(lambda row: round( 0.4*row['English'] + 0.3*row['History'] + .3*row['Geography'] ) , axis=1)
df
```

Create a new column, called `nerd`, whose value is True if avgSc > avgA and False otherwise

```python
# write your code here
df['nerd'] = df.apply(lambda row: True if row['avgSc'] > row['avgA'] else False, axis = 1)
df
```

Remove rows where History is less than English

```python
drop_index = df[df['History']<df['English']].index
df.drop(index = drop_index)
```

Finally, remove avgSc and avgA columns

```python
# write your code here
df.drop(columns= ['avgA','avgSc'])
```

## Exercise 5. Data Manipulation (My first Search Engine!)

We have a document collection (a list of strings).

Create a DataFrame from this list.

```python
docs= ["Sugar is bad to consume",
       "My sister likes to have sugar but not my father", 
       "My father spends a lot of time driving my sister around to dance practice",
       "Doctors suggest that driving may cause increased stress and blood pressure",
       "Sometimes I feel pressure to perform well at school", 
       "my father never seems to drive my sister to do better",
       "Health experts say that Sugar is not good for your lifestyle"]
# write your code here
df = pd.DataFrame(docs, columns = ['text'])
df
```

Use `reset_index()` to make the index as a column, and rename this new column from `index` to `docid`

```python
# write your code here
df.reset_index(inplace = True)
df.rename(columns = {'index': 'docid'}, inplace = True)
df
```

Now, create an empty list, called `inv_index`,
Then, iterate over the DataFrame rows (as tuples), and for each text:
* convert into lower case, 
* split it into words, and for each word:
    * make a tuple of (word, index), and
    * append it to `inv_index` list
Don't print the list, just print its length    

```python
# write your code here
inv_index = []
for idx, docid, text in df.itertuples():
  row_text = text.lower().split()
  inv_index.extend([(val, idx) for val in row_text])
print(len(inv_index))
```

Create a DataFrame, called `df_inv`, from the list `inv_index`, with column names: `Word`, and `Document`.
Print its shape, and show the head.

```python
# write your code here
df_inv = pd.DataFrame(inv_index, columns =['Word', 'Document'])
print(df_inv.shape)
df_inv.head()
```

Remove (inplace) from df_inv the rows where the word belongs to the following list of stop-words

```python
stopwords = ['a', 'and', 'at', 'but', 'do', 'for', 'have', 
             'i', 'is', 'my', 'not', 'of', 'that', 'to', 'your']
```

```python
# write your code here
df_inv = df_inv.loc[~df_inv['Word'].isin(stopwords)]
df_inv.head()
```

Write a function, called `search`, that takes 1 argument called `query`:
* this function should select from `df_inv` the rows where `Word` equals `query`
* get the `Document` column of those rows, call it: `res_docs` 
* then find in the DataFrame `df` the rows where the column `docid` is in `res_docs`
* and return the `text` of those rows (convert it to a list)

```python
def search(query):
    # write your code here
    res_docs = df_inv.loc[df_inv['Word'] == query, 'Document'].unique()
    return list(df.loc[df['docid'].isin(res_docs), 'text'])
```

Now, use this function to search for the word `'sugar'`

```python
search('sugar')
```
