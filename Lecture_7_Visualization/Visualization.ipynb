```python
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

plt.bar(pos, popularity, align='center')
plt.xticks(pos, languages)
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math & Data \nby % Popularity on Stack Overflow');
```

http://stackoverflow.com/research/developer-survey-2016#most-popular-technologies-per-occupation

```python
# change the bar color to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# change one bar, the python bar, to a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
# remove the Y label since bars are directly labeled
#plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', 
          alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=False, left=False, right=False, 
                labelleft=False, labelbottom=True)

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
# direct label each bar with Y axis values
for bar in bars:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, 
                   str(int(height)) + '%', ha='center', color='w', fontsize=11)
```

`plt.gca()` --> **g**et **c**urrent **a**xis

## Sub-plots

```python
df = pd.read_csv('iris.csv')
print(df.shape)
df.head()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    axs[0].scatter(df_['petal_length'], df_['petal_width'], label=spec)
axs[0].set_xlabel('Petal Length')
axs[0].set_ylabel('Petal Width')
axs[0].legend();

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    axs[1].scatter(df_['petal_length'], df_['sepal_length'], label=spec)
axs[1].set_xlabel('Petal Length')
axs[1].set_ylabel('Sepal Length')
axs[1].legend();
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    axs[0].scatter(df_['petal_length'], df_['petal_width'], label=spec)
    axs[1].scatter(df_['petal_length'], df_['sepal_length'], label=spec)
    
axs[0].set_xlabel('Petal Length')
axs[0].set_ylabel('Petal Width')
axs[0].legend();

axs[1].set_xlabel('Petal Length')
axs[1].set_ylabel('Sepal Length')
axs[1].legend();
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    ax1.scatter(df_['petal_length'], df_['petal_width'], label=spec)
    ax2.scatter(df_['petal_length'], df_['sepal_length'], label=spec)
    
ax1.set_xlabel('Petal Length')
ax1.set_ylabel('Petal Width')
ax1.legend();

ax2.set_xlabel('Petal Length')
ax2.set_ylabel('Sepal Length')
ax2.legend();
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey='row')

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    ax1.scatter(df_['petal_length'], df_['petal_width'], label=spec)
    ax2.scatter(df_['petal_length'], df_['sepal_length'], label=spec)
    
ax1.set_xlabel('Petal Length')
ax1.set_ylabel('Petal Width')
ax1.legend();

ax2.set_xlabel('Petal Length')
ax2.set_ylabel('Sepal Length')
ax2.legend();
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8), sharex='col')

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    ax1.scatter(df_['petal_length'], df_['petal_width'], label=spec)
    ax2.scatter(df_['petal_length'], df_['sepal_length'], label=spec)
    
ax1.set_xlabel('Petal Length')
ax1.set_ylabel('Petal Width')
ax1.legend();

ax2.set_xlabel('Petal Length')
ax2.set_ylabel('Sepal Length')
ax2.legend();
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))

for spec in df['species'].unique():
    df_ = df[df['species']==spec]
    ax1.scatter(df_['petal_length'], df_['petal_width'], label=spec)
ax1.set_xlabel('Petal Length')
ax1.set_ylabel('Petal Width')
ax1.legend();

c = 'sepal_length'
data = [df1[c], df2[c], df3[c]]
ax2.boxplot(data);
ax2.set_title(c)
ax2.set_xticks([1,2,3])
ax2.set_xticklabels(specs);

y1 = np.random.randint(10,30, size=10)
y2 = np.random.randint(20,40, size=10)
x = np.arange(len(y1))
ax3.plot(x, y1, c='r', marker='s', linestyle=':', label='Class 1')
ax3.plot(x, y2, c='b', marker='*', linestyle='-', label='Class 2')
ax3.set_xlabel('Length')
ax3.set_ylabel('Size')
ax3.set_xticks(np.arange(0,11))
ax3.set_yticks(np.arange(0,41, 5))
ax3.legend();

ax4.bar(pos, popularity, align='center')
ax4.set_xticks(pos, languages)
ax4.set_ylabel('% Popularity')
ax4.set_title('Top 5 Languages by % Popularity');

plt.savefig('mixed_drawing.png', dpi=200, bbox_inches='tight')
```
