## Built-in methods

You can easily create an array filled with 0’s or with 1’s. 
* The function `np.zeros()` creates an array full of zeros. 
* The function `np.ones()` creates an array full of ones. 
* By default, the dtype of the created array is float64.

### `np.zeros()`

```python
np.zeros(3)
```

```python
np.zeros((4,5))
```

```python
np.zeros((3, 2), dtype=int)
```

### `np.ones()`

```python
np.ones(4)
```

```python
np.ones((2,3))
```

### `np.eye()`

Creates an identity matrix:
* squared ($n \times n$)
* `1` at the diagonal
* `0` elsewhere

```python
np.eye(4)
```

```python
np.eye(4,dtype = int)
```

### `np.full()`

Creates a full matrix

```python
np.full(shape=4, fill_value=3)
```

```python
np.full((2,3), 5)
```

### `np.arange()`

Return evenly spaced values within a given interval.

```python
# similar to range function
np.arange(4)
# np.arange(0, 4, 1)
```

```python
np.arange(2, 7)
```

```python
np.arange(30, 101, 10)
```

### `np.linspace()`

```python
# linspace (start, end, no of entries) start and end are both included endpoint = true or false and depending on it it will stop at 11 or 10
np.linspace(0, 11, 3,endpoint = False, retstep = True)
```

```python
np.linspace(0, 10, 5) #10/4
```

Rounding

Exponents and logarithms

Arithmetic operations
* `reciprocal`, `power`, `sqrt`, `cbrt`, `square`, `absolute`
* `add`, `subtract`, `multiply`, `divide`

 Example

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100) 
f1 = np.sin(x)
f2 = np.cos(x)
plt.plot(x, f1, label='sin')
plt.plot(x, f2, label='cos')
plt.legend();
```

## Random Numbers

* `rand`: Create an array of the given shape and populate it with random samples from a **uniform** distribution over [0, 1).

* `randn`: Return a sample (or samples) from the **"standard normal"** distribution. Unlike rand which is uniform

* `randint`: Return random **integers** from low (inclusive) to high (exclusive).


### `rand`
 
Create an array of the given shape and populate it with random samples from a **uniform** distribution over [0, 1).

```python
np.random.rand(4)
```

```python
np.random.rand(5,5)
```

### `randn`
 
Return a sample (or samples) from the **"standard normal"** distribution (mean=0, std=1). 
Unlike rand which is uniform.

```python
np.random.randn(4)
```

```python
np.random.randn(3, 4)
```

### `randint`

Return random integers from `low` (inclusive) to `high` (exclusive).

`np.random.randint(low, high, size)`

```python
np.random.randint(1, 100)
```

```python
np.random.randint(1, 100, 5)
```

```python
np.random.randint(20, 50, (4,3))
```

```python
a = np.random.rand(1000)
plt.hist(a);
```

```python
a = np.random.randn(1000)
plt.hist(a);
```

To obtain *uniform* random numbers in the range $[a,b)$ (between a and b):

`a + np.random.rand() * (b-a)`

```python
# uniform random numbers in [20, 80)
X = 20 + np.random.rand(1000) * 60
plt.hist(X);
plt.savefig('uniform_20_80.png', dpi=200)
```

```python
X[:5]
```

```python
X.mean(), X.std()
```

To obtain *normal* random numbers with mean $m$ and standard deviation $s$:

`m + np.random.rand() * s`


```python
# uniform random numbers with mean $m=50$ and std $s=10$
X = 50 + np.random.randn(1000) * 10
plt.hist(X);
```

```python
X.mean(), X.std()
```
